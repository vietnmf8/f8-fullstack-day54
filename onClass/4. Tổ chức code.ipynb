```python
// autoClearRevokedToken.js
const revokedTokenModel = require("@/models/revokedToken.model");

async function autoClearRevokedToken() {
    const result = await revokedTokenModel.deleteRevokeToken();
    console.log(`Đã xoá ${result} token hết hạn!`);
}

module.exports = autoClearRevokedToken;
```

```python
// onClass\schedule.js
require("dotenv").config();
require("module-alias/register");
const autoClearRevokedToken = require("@/schedulers/autoClearRevokedToken");
const { CronJob } = require("cron");

/**
 * 1h sáng mỗi ngày
 * Tự động Clear những token đã nằm trong Blacklist
 */

new CronJob("0 1 * * *", autoClearRevokedToken).start();
```
