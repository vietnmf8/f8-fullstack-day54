```python
// Bước 1: Tạo onClass\schedule.js

import { CronJob } from 'cron';

const job = new CronJob(
	'* * * * * *', // cronTime
	function () {
		console.log('You will see this message every second');
	}, // onTick
	null, // onComplete
	true, // start
	'America/Los_Angeles' // timeZone
);
// job.start() is optional here because of the fourth parameter set to true.

/* 
* cronTime (Tham số 1): 
	Đây là chuỗi 6 dấu sao. Nó quyết định "Tần suất" công việc. Trong ví dụ này là * * * * * * (Mỗi giây một lần).

* onTick (Tham số 2): 
	Đây là "Trái tim" của Job. Nó là một hàm (callback function) chứa toàn bộ logic em muốn thực thi khi đúng giờ. Ví dụ: gọi Model để xóa dữ liệu, gửi email, v.v.

* onComplete (Tham số 3): 
	Hàm này sẽ tự động chạy khi Job bị dừng lại (gọi lệnh .stop()). Thường dùng để dọn dẹp bộ nhớ hoặc log lại trạng thái kết thúc. Nếu không dùng, em cứ để null.

* start (Tham số 4): Kiểu dữ liệu Boolean (true/false):
	Nếu để true: Job sẽ tự kích hoạt ngay khi dòng code này được đọc.
	Nếu để false: Em phải gọi thêm lệnh job.start() thì nó mới bắt đầu "đếm ngược".

* timeZone (Tham số 5): 
	- Xác định múi giờ để đối chiếu. Nếu em làm app cho người Việt, hãy luôn để 'Asia/Ho_Chi_Minh' để tránh việc code chạy lúc 3h sáng (giờ Mỹ) thay vì 8h sáng (giờ Việt).
*/
```

```python
/* 
Hãy tưởng tượng em đang đặt một Lệnh sản xuất lính trong game chiến thuật:
	- cronTime: Cứ 30 giây sản xuất 1 đợt.
	- onTick: Xuất hiện 3 con lính cầm kiếm.
	- start: Vừa vào trận là tự sản xuất luôn (true).
	- timeZone: Theo giờ của máy chủ Đông Nam Á.
*/
```
